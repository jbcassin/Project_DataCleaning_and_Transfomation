```python
import pandas as pd
import re
```

```python
df = pd.read_csv("Uncleaned_DS_jobs.csv", index_col=0)
df.head()
```

```python
df.info()
```

```python
# Extraction des salaires minimaux et maximaux
def extract_salary_range(salary_estimate):
    salary_parts = re.findall(r'\d+', salary_estimate)
    min_salary = int(salary_parts[0])
    max_salary = int(salary_parts[1])
    return min_salary, max_salary

# Appliquer la fonction d'extraction et créer les colonnes min salary et max salary
df[['Min Salary', 'Max Salary']] = df['Salary Estimate'].apply(extract_salary_range).apply(pd.Series)
```

```python
df.head()
```

```python
df["Job Description"] 
```

```python
# Suppression des cinq derniers caractères de la colonne "Company Name"
df['Company Name'] = df['Company Name'].apply(lambda x: x[:-4])
```

```python
df["Average Salary"] = (df["Min Salary"] + df["Max Salary"]) /2
```

```python
df
```

```python
# Création de la colonne State
df['State'] = df['Location'].str[-2:]
df['Location'] = df['Location'].apply(lambda x: x[:-4])
```

```python
df.head()
```

```python
# Création de la colonne Company Age
from datetime import datetime
current_year = datetime.now().year
df['Company Age'] = current_year - df['Founded']
```

```python
# Text mining

from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

# Tokenization des descriptions de poste
df['Tokenized Job Description'] = df['Job Description'].apply(lambda text: word_tokenize(text, language='english'))
```

```python
df['Tokenized Job Description']
```

```python
# Importer stopwords de la classe nltk.corpus
from nltk.corpus import stopwords

# Initialiser la variable des mots vides
stop_words = set(stopwords.words('english'))

# Définir la fonction stop_words_filtering
def stop_words_filtering(mots):
    tokens = [mot for mot in mots if mot.lower() not in stop_words]  # Utilisez mot.lower() pour comparer avec les mots vides en minuscules
    return tokens

# Appliquer la fonction de filtrage des mots vides
df['Filtered Job Description'] = df['Tokenized Job Description'].apply(stop_words_filtering)
```

```python
df['Filtered Job Description']
```

```python
# tokenisation

#Importer la pckage nécessaire
from nltk.tokenize.regexp import RegexpTokenizer

# Extraire les tokens en utilisant le RegexpTokenizer
tokenizer = RegexpTokenizer(r'[a-zA-Zé]{4,}')  # Ceci correspond aux mots de 4 caractères ou plus
df['Extracted Tokens'] = df['Filtered Job Description'].apply(lambda mots: tokenizer.tokenize(' '.join(mots).lower()))


# Afficher les tokens
print(df['Extracted Tokens'])
```

```python
# Word Cloud

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Créer une chaîne de caractères à partir des tokens extraits
all_tokens = ' '.join(' '.join(tokens) for tokens in df['Extracted Tokens'])

# Créer le nuage de mots
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_tokens)

# Afficher le nuage de mots
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Créer la colonne "Python" avec la valeur 1 si "python" est présent dans les tokens extraits
df['Python'] = df['Extracted Tokens'].apply(lambda tokens: 1 if 'python' in tokens else 0)
# Créer la colonne "Python" avec la valeur 1 si "python" est présent dans les tokens extraits
df['Excel'] = df['Extracted Tokens'].apply(lambda tokens: 1 if 'excel' in tokens else 0)
# Créer la colonne "Python" avec la valeur 1 si "python" est présent dans les tokens extraits
df['Hadoop'] = df['Extracted Tokens'].apply(lambda tokens: 1 if 'hadoop' in tokens else 0)
# Créer la colonne "Python" avec la valeur 1 si "python" est présent dans les tokens extraits
df['Spark'] = df['Extracted Tokens'].apply(lambda tokens: 1 if 'Spark' in tokens else 0)
# Créer la colonne "Python" avec la valeur 1 si "python" est présent dans les tokens extraits
df['AWS'] = df['Extracted Tokens'].apply(lambda tokens: 1 if 'aws' in tokens else 0)
# Créer la colonne "Python" avec la valeur 1 si "python" est présent dans les tokens extraits
df['Tableau'] = df['Extracted Tokens'].apply(lambda tokens: 1 if 'tableau' in tokens else 0)
# Créer la colonne "Python" avec la valeur 1 si "python" est présent dans les tokens extraits
df['Cloud'] = df['Extracted Tokens'].apply(lambda tokens: 1 if 'Cloud' in tokens else 0)


```

```python
# Nouvelle colonne "Job_simp"

# Définir la fonction pour simplifier les titres d'emploi
def simplify_job_title(job_title):
    if 'Data Scientist' in job_title:
        return 'Data Scientist'
    elif 'Data Analyst' in job_title:
        return 'Data Analyst'
    elif 'Data Engineer' in job_title:
        return 'Data Engineer'
    elif 'Machine Learning Engineer' in job_title:
        return 'Machine Learning Engineer'
    else:
        return 'Other'

# Appliquer la fonction pour créer la colonne "Job_simp"
df['Job_simp'] = df['Job Title'].apply(simplify_job_title)
```
